```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(Path('../Resources/oil_gweekly.csv'))
df.head()
df = df.dropna()
```

```python
oildf= df[['my','week','value_x','value_y']]
#oil_gdf = oil_gdf[['my',  'value','year']]
```

```python
plt.scatter(oildf.value_x, oildf.value_y)
plt.xlabel('oil price$')
plt.ylabel('gas price$')
plt.show()
```

```python
X = oildf.value_x.values.reshape(-1, 1)
```

```python
X[:10]
```

```python
X.shape
```

```python
y = oildf.value_y
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X, y, s=10, marker='x')
plt.plot(X, y_pred, color='red')
plt.xlabel('oil price ($)')
plt.ylabel('gas price ($)')
# plt.xlim(0, 100)
# plt.ylim(0, 5)
# plt.figure(figsize=(20, 20), dpi=80)
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
    y, test_size=.3, random_state=0)
```

# split  

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
#training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```
