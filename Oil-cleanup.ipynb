```python
# Clean up Oil dataset
# Final Project
import pandas as pd
file_path = "./Resources/oil-prices.csv"
oil_df = pd.read_csv(file_path)
file_path = "./Resources/pswrgvwallfull.csv"
pwg_df = pd.read_csv(file_path)
file_path = "./Resources/2016global_crisis.csv"
gcr_df = pd.read_csv(file_path)

print(oil_df.head())
print(pwg_df.head())
print(gcr_df.head())
```

```python
gcr_df.head()
```

```python
# Drop null and na for dataframe
gcr_df = gcr_df.dropna()
gcr_df = gcr_df[['year','crisis','exch_usd']]
gcr_df.head()
```

```python
# Drop null and na for dataframe
oil_df = oil_df.dropna()
oil_df.head()
```

```python
# data types to rename columns
oil_df.dtypes
```

```python
# rename columns
oil_df.rename(columns={"Macrotrends Data Download":"date", "Unnamed: 1":"value"})
```

```python
pwg_df.rename(columns={"Macrotrends Data Download":"date", "Unnamed: 1":"value"})
#oil_df.set_index('date')
```

```python
pwg_df = pwg_df.iloc[1:]
oil_df = oil_df.iloc[1:]
```

```python
pwg_df.head()
```

```python
 # A.merge(B, left_on='lkey', right_on=['date', 'year'], how='outer')
# pandas.merge(left, right, how='inner',
oilgas_df = pd.merge(oil_df, pwg_df, on=['date'], how='outer' )
oilgas_df.head()
```

```python
# Saving joined cleaned data
file_path = "./Resources/oil_clean.csv"
oil_df.to_csv(file_path, index=False)
```
